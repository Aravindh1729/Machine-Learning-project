
Converting all catagoric values to numerical values

```python
from sklearn.preprocessing import LabelEncoder
columns_to_encode = ['TRAIN_TYPE', 'ORIGIN','DESTINATION','TRAIN_CLASS']
le = LabelEncoder()#handling catagorical values
for column in columns_to_encode:
    df[column] = le.fit_transform(df[column])
```

Taking values for x and y. x represents the features or independent variables of the dataset. y represents the target variable or dependent variable

```python
x=df.drop(['PRICE','ROUTE','INSERT_DATE','FARE'],axis=1)
y=df['PRICE']
```

Importing train test split

```python
from sklearn.model_selection import train_test_split
```

Spliting the dataset into train and test


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

Checking for the shape of train and test for both independent and dependant variable.


```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

Importing r2 -statistical measure indicating the goodness of fit of a regression model.

```python
from sklearn.metrics import r2_score
```

Linear regression

```python
from sklearn.linear_model import LinearRegression    # shows linear relationship between a dependent (y) and one or more independent (y) variables
```

Creation of linear regresssion

```python
modelmlr = LinearRegression()
```

Training the model

```python
modelmlr.fit(x_train, y_train)
```

Prediction on test data

```python
y_pred = modelmlr.predict(x_test)
```

Computing the scores

```python
scoretrain = round(modelmlr.score(x_train, y_train) * 100, 2)
scoretest=round(modelmlr.score(x_test,y_test)*100,2)
r2=round(r2_score(y_test,y_pred)*100,2)
```

```python
print('The score of train is',scoretrain)
print('The score of test is',scoretest)
print('The r2 score is',r2)
```
